```python
#trabajar dataframes
import pandas as pd
#nube de palabras
from stylecloud import gen_stylecloud
from PIL import Image
#preprocesamiento
import nltk
from nltk import SnowballStemmer
import spacy
#modelado
from sklearn.model_selection import train_test_split
from nltk import NaiveBayesClassifier
from nltk import classify
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from keras.layers import Dense
from keras.callbacks import EarlyStopping, ModelCheckpoint
import unicodedata
import re
import keras_tuner as kt
import tensorflow as tf
import random
import numpy as np
import tensorflow as tf
from sklearn.metrics import confusion_matrix
random.seed(1234)
np.random.seed(1234)
tf.random.set_seed(1234)
```

```python
#creamos nubes de palabras
def get_wordcloud(text, icon="fas fa-comments", background_color=None, output_name="./wordcloud.png"):
    gen_stylecloud(text=text, icon_name=icon, background_color=background_color, output_name=output_name)
    return Image.open(output_name)
```

```python
#pasamos a minúscula
def clean_text(text, pattern="[^a-zA-Z0-9 ]"):
    cleaned_text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
    cleaned_text = re.sub(pattern, " ", cleaned_text.decode("utf-8"), flags=re.UNICODE)
    cleaned_text = u' '.join(cleaned_text.lower().split())
    return cleaned_text
```

```python
#quitamos stopwords y signos de puntuación
def normalize(text):
    doc = nlp(text)
    words = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
    return words
```

```python
#estematizamos
def estematizar(text):
    return [spanishstemmer.stem(token) for token in text]
```

```python
#formato que requiere ML
def formato(lista):
    temp=[[x,True] for x in lista]
    return dict(temp)
```

```python
#lectura de datos
df=pd.read_csv('final(3).csv')
df.columns=['mensaje','sentimiento']
```

```python
#limpiamos y guardamos los datos limpios
df=df.drop(df[df['sentimiento']=='sentimiento'].index)
df.mensaje=df.mensaje.str.replace('\n','')
df.to_csv('final2.csv',index=False)
```

```python
df=pd.read_csv('final2.csv')
df.columns=['mensaje','sentimiento']
df=df.drop(df[df['sentimiento']<-1].index)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
#limpiamos nuestros datos con las funciones
spanishstemmer=SnowballStemmer('spanish')
nlp = spacy.load('es_core_news_sm')
df["mensaje"] = df["mensaje"].map(clean_text)
df['mensaje']=df['mensaje'].apply(normalize)
df=df.drop(df[df.mensaje.apply(len)==0].index)
```

```python
df['mensaje']=df['mensaje'].apply(estematizar)
df.head()
```

```python
df['mensajeML']=df.mensaje.apply(formato)
```

```python
MAX_NB_WORDS = 3000
MAX_SEQUENCE_LENGTH = 15
EMBEDDING_DIM = 3
tokenizer = Tokenizer(num_words=MAX_NB_WORDS)
```

```python
tokenizer.fit_on_texts(df['mensaje'].values)
word_index = tokenizer.word_index
X = tokenizer.texts_to_sequences(df['mensaje'].values)
```

```python
X = pad_sequences(X, maxlen=MAX_SEQUENCE_LENGTH)
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(df["sentimiento"])
```

```python
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, shuffle=True,stratify=y)
```

```python
X.shape[1]
```

```python
#hiperparametrización de la red neuronal
def model_builder(hp):
    
    model=tf.keras.Sequential()
    model.add(tf.keras.layers.Embedding(3000,3,input_length=15))
    #MAX_NB_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
    model.add(tf.keras.layers.SpatialDropout1D(0.2))
    
    
    hp_activation=hp.Choice('activation',values=['relu','tanh'])
    hp_activation2=hp.Choice('activation2',values=['sigmoid','softmax'])
    hp_activation3=hp.Choice('activation3',values=['sigmoid','softmax'])
    hp_layer_1=hp.Int('layer_1',min_value=1,max_value=10,step=2)
    hp_layer_2=hp.Int('layer_2',min_value=1,max_value=10,step=2)
    hp_learning_rate=hp.Choice('learning_rate',values=[1e-2,1e-3,1e-4])
    
    
    model.add(tf.keras.layers.LSTM(units=hp_layer_1,activation=hp_activation))
    model.add(tf.keras.layers.Dense(units=hp_layer_2, activation=hp_activation2))
    model.add(tf.keras.layers.Dense(3, activation=hp_activation3))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate),
                 loss=tf.keras.losses.CategoricalCrossentropy(),
                 metrics=['accuracy'])
    return model 
```

```python
tuner=kt.Hyperband(model_builder,objective='val_accuracy',max_epochs=1000,factor=3,directory='dir',
                   project_name='z',seed=4321)#x fue el primero
stop_early=tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=15, min_delta=0.0000001)
tuner.search(X_train,y_train,epochs=100,validation_split=0.2,callbacks=[stop_early],batch_size=32)
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_hps.values
```

```python
#entrenamos NN
model = tuner.hypermodel.build(best_hps)
history = model.fit(X_train, y_train, epochs=1000, validation_split=0.2,
                    callbacks=[stop_early],batch_size=32)
```

```python
loss, acc = model.evaluate(X_train, y_train)
loss, acc2 = model.evaluate(X_test, y_test)
y_p = model.predict(X_test)
yNN_p=[le.inverse_transform([np.argmax(i, axis=None, out=None)])[0] for i in y_p]
yNN_test=[le.inverse_transform([np.argmax(i, axis=None, out=None)])[0] for i in y_test]
cm = confusion_matrix(yNN_test,yNN_p)
```

```python
#VP FP
#FN VN
XML=df.mensajeML
yML=df.sentimiento
XML_train, XML_test, yML_train, yML_test = train_test_split(XML, yML, test_size=0.2, random_state=42,stratify=yML)
classifier = NaiveBayesClassifier.train(list(zip(XML_train,yML_train)))
print("Accuracy ML entrenamiento:", classify.accuracy(classifier, list(zip(XML_train,yML_train))))
print("Accuracy ML prueba:", classify.accuracy(classifier, list(zip(XML_test,yML_test))))
y_pML = XML_test.apply(classifier.classify)
cm2 = confusion_matrix(yML_test,y_pML)
print('Matriz de confusión ML')
print(cm2)
```

```python
print("Accuracy NN entrenamiento:", acc)
print("Accuracy NN prueba:", acc2)
print('Matriz de confusión NN')
print(cm)
```
